###  Finding out the Missing Data in first 15 coulmns  

```python
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
```

```python
missing_data(df).head(15)
```

```python
import seaborn as sns
sns.set_context("paper", font_scale = 2, rc = {"font.size": 20,"axes.titlesize": 25,"axes.labelsize": 20})   
b = sns.catplot(data = bf, kind = 'count', x = 'Family',aspect=3,  order = bf['Family'].value_counts().index)
plt.title(' Loans taken according  number members in the  Family ')
b.set_xticklabels(rotation = 90)
plt.show()
```

```python
df[['loan_amnt', 'emp_length']].groupby(['emp_length']).mean().sort_values("loan_amnt", ascending = False).head(20).plot.bar()
plt.title(' Loan amount taken as per experience   ')
plt.show()
```

```python
df[['int_rate', 'emp_title']].groupby(['emp_title']).max().sort_values("int_rate", ascending = False).head(20).plot.bar()
plt.title('Intrest as per Job title ')
plt.show()
```
